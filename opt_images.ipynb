## Modificar el peso de varias imágenes en un directorio

```python
from PIL import Image
#Módulo para manejar directorios
import os

# Directorios de entrada y salida
input_directory = './CAPUTURAS ONDEMAND'
output_directory = './CAPUTURAS ONDEMAND OPT no RZS'

# Función para optimizar una imagen
def optimize_image(input_path, output_path):
    image = Image.open(input_path)
    image.save(output_path, optimize=True)

# Función para recorrer el directorio y optimizar imágenes
def optimize_images_in_directory(input_dir, output_dir):
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                input_path = os.path.join(root, file)
                output_path = input_path.replace(input_dir, output_dir)
                # Agregar '_opt' al nombre del archivo
                output_path = output_path[:-4] + '_opt' + output_path[-4:]
                # Crear directorios si no existen
                os.makedirs(os.path.dirname(output_path), exist_ok=True)
                # Función creada anteriormente
                optimize_image(input_path, output_path)
                # Msj de confirmación
                print(f'Imagen optimizada: {input_path} -> {output_path}')

# Llamada a la función para optimizar imágenes en el directorio
optimize_images_in_directory(input_directory, output_directory)
```

## Modificar el peso y el tamaño (dimensiones) de varias imágenes en un directorio

```python
from PIL import Image
import os

# Directorios de entrada y salida
input_directory = './CAPUTURAS ONDEMAND'
output_directory = './CAPUTURAS ONDEMAND OPT RZS'

# Función para optimizar una imagen reduciendo su tamaño y calidad
def optimize_image(input_path, output_path):
    image = Image.open(input_path)

    # Reducción de dimensiones a la mitad
    width, height = image.size
    new_size = (width // 2, height // 2)
    resized_image = image.resize(new_size, Image.ANTIALIAS)

    # Optimización de calidad y peso
    resized_image.save(output_path, quality=50, optimize=True)

# Función para recorrer el directorio y optimizar imágenes
def optimize_images_in_directory(input_dir, output_dir):
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                input_path = os.path.join(root, file)
                output_path = input_path.replace(input_dir, output_dir)
                os.makedirs(os.path.dirname(output_path), exist_ok=True)  # Crear directorios si no existen

                #output_path = os.path.splitext(output_path)[0] + '_opt' + os.path.splitext(output_path)[1]
                output_path = os.path.splitext(output_path)[0] + os.path.splitext(output_path)[1]
                optimize_image(input_path, output_path)

                print(f'Imagen optimizada: {input_path} -> {output_path}')

# Llamada a la función para optimizar imágenes en el directorio
optimize_images_in_directory(input_directory, output_directory)
```
